#### Day 4 - B
Search for the word "MAS" overlapping diagonally in the letter grid

```python
#Import Libraries and settings

settings = {
    "day": 4,
    "test_data": 0
}
```

```python
#Load Input
def load_input(settings):
    #Derrive input file name
    if settings["test_data"]:
        data_subdir = "test"
    else:
        data_subdir = "actual"

    data_fp = f"./../input/{data_subdir}/{settings["day"]}.txt"

    #Open and read the file
    with open(data_fp) as f:
        data = f.read().split('\n')

    #Get the dimensions of the input and include it in the output



    return {
        "data": data,
        "width": len(data[0]),
        "height": len(data)
    }

grid = load_input(settings)
```

```python
#loc is a tuple of input indexes (x, y)
#dir is a tuple of delta indexs (i.e. (0, 1) is downward by 1 step)

#Return location after applying dir
def apply_dir(loc, idx_dir):

    return (loc[0] + idx_dir[0], loc[1] + idx_dir[1])

#Convert compass direction into a dir tuple
def translate_step(compass_dir, mag=1):
    v = 0
    h = 0

    if "N" in compass_dir:
        v = -mag
    elif "S" in compass_dir:
        v = mag

    if "E" in compass_dir:
        h = mag
    elif "W" in compass_dir:
        h = -mag

    return (h, v)

#Get all valid directions for a given point in the grid (cannot exceed grid boundaries)
def valid_to_search(loc, grid, limb_len=1):

    #If the target limbs would go out of bound then not a valid start location
    if (
        loc[0] <= (limb_len - 1) or 
        loc[0] >= (grid["width"] - 1) - (limb_len - 1) or
        loc[1] <= (limb_len - 1) or 
        loc[1] >= (grid["height"] - 1) - (limb_len - 1)
    ):
        return False

    else:
        return True

#Check the diagonals for a given location
def check_diag(grid, loc, ori, targets=["M", "S"]):

    #Determine which orientation is being checked
    if ori == 0:
        limb_dirs = ["NW", "SE"]
    else:
        limb_dirs = ["NE", "SW"]

    #Get the first limb
    first_loc = apply_dir(loc, translate_step(limb_dirs[0]))
    limb_char_1 = grid["data"][first_loc[1]][first_loc[0]] 

    #If the first limb is one of the targets
    if limb_char_1 in targets:
        #Check if it's the first target found
        if limb_char_1 == targets[0]:
            second_target = targets[1]
        else:
            second_target = targets[0]

        #Get the second limb
        second_loc = apply_dir(loc, translate_step(limb_dirs[1]))
        limb_char_2 =  grid["data"][second_loc[1]][second_loc[0]]

        #Check if the second limb is the correct character
        if limb_char_2 == second_target:
            return True

    #Invalid diag
    return False

#Check the current provided location for the the target centre
def check_location(grid, loc, target="A"):
    #Check this is a valid starting location
    if valid_to_search(loc, grid):

        loc_value = grid["data"][loc[1]][loc[0]]

        #Check if potential target found
        if loc_value == target[0]:
            #Check both diagonals
            if check_diag(grid, loc, ori=0):
                if check_diag(grid, loc, ori=1):
                    return 1

    #Not a match
    return 0

#Function to traverse grid for words
def wordsearch(grid):

    #Running total
    words_found = 0
    
    #Iterate through the grid
    for idx_y in range(grid["height"]):
        for idx_x in range(grid["width"]):
            loc = (idx_x, idx_y)

            res = check_location(grid, loc)
            if res > 0:
                #print(f"{res} target(s) found at {loc}")
                pass

            words_found += res

    return words_found
```

```python
print(wordsearch(grid))
```
