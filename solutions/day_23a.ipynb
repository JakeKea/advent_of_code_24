#### Day 23 - A
The input is connections between computers.
Find the number of 3 way connections involving a computer beginning with t.

```python
#Import Libraries and settings

settings = {
    "day": 23,
    "test_data": 0
}
```

```python
#Load Input
def load_input(settings):
    #Derrive input file name
    if settings["test_data"]:
        data_subdir = "test"
    else:
        data_subdir = "actual"

    data_fp = f"./../input/{data_subdir}/{settings["day"]}.txt"

    #Open and read the file
    with open(data_fp) as f:
        lines = f.read().split('\n')

    #Build dictionary of connections
    connections = {}

    for line in lines:
        pcs = line.split("-")
        if pcs[0] not in connections.keys():
            connections[pcs[0]] = [pcs[1]]
        else:
            connections[pcs[0]].append(pcs[1])

        if pcs[1] not in connections.keys():
            connections[pcs[1]] = [pcs[0]]
        else:
            connections[pcs[1]].append(pcs[0])

    return connections

connections = load_input(settings)
```

```python
#Find all 3 way connections with computers starting with t
connections_3 = set()

t_hosts = [x for x in connections.keys() if x[0] == "t"]

for host in t_hosts:
    for guest in connections[host]:
        for third_party in connections[host]:
            if third_party in connections[guest]:
                new_connection = [host, guest, third_party]
                new_connection.sort()
                connections_3.add(tuple(new_connection))
```
