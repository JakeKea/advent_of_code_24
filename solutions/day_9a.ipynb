Decode line of block files, space between them and IDs for each block file.

Compress the decoded line and get the sum of products of the block id and index.

```python
#Import Libraries and settings

settings = {
    "day": 9,
    "test_data": 0
}
```

```python
#Load Input
def load_input(settings):
    #Derrive input file name
    if settings["test_data"]:
        data_subdir = "test"
    else:
        data_subdir = "actual"

    data_fp = f"./../input/{data_subdir}/{settings["day"]}.txt"

    #Open and read the file
    with open(data_fp) as f:
        line = f.read()

    pairs = []
    for pair in range(int(len(line)/2)):
        pairs.append((line[pair*2], line[pair*2+1]))
    pairs.append((line[-1], 0))

    return pairs

line_raw = load_input(settings)
```

```python
#Decode the line by expressing it in terms on file blocks and memory location
def decode_line(line):
    decoded_line = ""
    for idx, pair in enumerate(line):
        #Add the next file block
        #Convert the idx to a utc char so it can be represented by 1 character
        decoded_line += chr(idx+65) * int(pair[0])
        #Add the next space between files
        decoded_line +=  "." * int(pair[1])

    return decoded_line

line_decoded = decode_line(line_raw)
```

```python
print(line_decoded)
```

```python
#Compress the line by filling the gaps between files
def compress_line(line):
    #Count of num in line
    num_count = len(line) - line.count(".")
    print(num_count)
    
    #Iterate through line until all numbers processed
    compressed_line = ""
    back_inserts = 0
    for idx in range(num_count):

        #If the next char is a num, keep
        if line[idx] != ".":
            compressed_line += line[idx]
        #Else take the rightmost num in the line
        else:
            #Iterate until last character is not a "."
            last_dot = line[-1] == "."
            while(last_dot):
                line = line[:-1]
                last_dot = line[-1] == "."
            compressed_line += line[-1]
            line = line[:-1]

    return compressed_line

line_out = compress_line(line_decoded)
```

```python
#Get the output from the compressed line
def calc_out(line_out):
    tens = 0
    subtotal = 0
    for idx, num in enumerate(line_out):
        subtotal += idx*(ord(num)-65)

    return subtotal
print(calc_out(line_out))
```
