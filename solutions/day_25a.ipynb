How many valid combinatons of keys and locks are there?

```python
#Import Libraries and settings

settings = {
    "day": 25,
    "test_data": 0
}
```

```python
#Load Input
def load_input(settings):
    #Derrive input file name
    if settings["test_data"]:
        data_subdir = "test"
    else:
        data_subdir = "actual"

    data_fp = f"./../input/{data_subdir}/{settings["day"]}.txt"

    #Open and read the file
    with open(data_fp) as f:
        lines = f.read().split('\n')

    

    line_idx = 0
    blocks = [[]]

    #Convert input into blocks
    for line in lines:
        if line == "":
            line_idx += 1
            blocks.append([])
            
        else:
            blocks[line_idx].append(line)

    #Seperate out locks and keys
    locks = []
    keys = []

    for block in blocks:
        if block[0][0] == "#":
            locks.append(block)
        else:
            keys.append(block)


    return locks, keys

locks, keys = load_input(settings)
print(keys)
```

```python
#Conert 2D block array into array of dimensions
def get_dimensions(block, lock=True, width=5):

    #Work on copy of block
    block = block[:]

    #Initialise dims array
    dims = [0] * width

    #Convert key to lock so code can be reused
    if not lock:
        block = block[::-1]

    #Check if each caharacter is # and update dims accordingly
    for line in block[1:]:
        for idx_col, char in enumerate(line):
            if char == "#":
                dims[idx_col] += 1

    return dims
```

```python
lock_dims = []
key_dims = []

for lock in locks:
    lock_dims.append(get_dimensions(lock))

for key in keys:
    key_dims.append(get_dimensions(key, lock=False))
```

```python
def validiate_combination(key, lock, height=5):
    for idx in range(len(key)):
        if key[idx] + lock[idx] > height:
            return False
    return True
```

```python
combinations = 0

print(len(keys), "keys")
print(len(locks), "locks")

for idx, key in enumerate(key_dims):

    for lock in lock_dims:
        if validiate_combination(key, lock):
            combinations += 1
    
    if idx % 10 == 0:
        print(idx)

print(combinations)
```
