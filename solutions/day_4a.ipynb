#### Day 4 - A
Search for the word "XMAS" in the letter grid

```python
#Import Libraries and settings

settings = {
    "day": 4,
    "test_data": 0
}
```

```python
#Load Input
def load_input(settings):
    #Derrive input file name
    if settings["test_data"]:
        data_subdir = "test"
    else:
        data_subdir = "actual"

    data_fp = f"./../input/{data_subdir}/{settings["day"]}.txt"

    #Open and read the file
    with open(data_fp) as f:
        data = f.read().split('\n')

    #Get the dimensions of the input and include it in the output



    return {
        "data": data,
        "width": len(data[0]),
        "height": len(data)
    }

grid = load_input(settings)
```

```python
#loc is a tuple of input indexes (x, y)
#dir is a tuple of delta indexs (i.e. (0, 1) is downward by 1 step)

#Return location after applying dir
def apply_dir(loc, idx_dir):

    return (loc[0] + idx_dir[0], loc[1] + idx_dir[1])

#Convert compass direction into a dir tuple
def translate_step(compass_dir, mag=1):
    v = 0
    h = 0

    if "N" in compass_dir:
        v = -mag
    elif "S" in compass_dir:
        v = mag

    if "E" in compass_dir:
        h = mag
    elif "W" in compass_dir:
        h = -mag

    return (h, v)

#Get all valid directions for a given point in the grid (cannot exceed grid boundaries)
def get_valid_dirs(loc, grid, target_len=4):

    all_dirs = ["N", "NE", "E", "SE", "S", "SW", "W", "NW"]
    invalid_dirs = []

    #At the top or bottom of the grid
    if loc[1] <= target_len - 2:
        invalid_dirs.append("N")
    elif loc[1] >= (grid["height"]-1) - (target_len-2):
        invalid_dirs.append("S")

    #At the top or bottom of the grid
    if loc[0] <= target_len - 2:
        invalid_dirs.append("W")
    elif loc[0] >= (grid["width"]-1) - (target_len-2):
        invalid_dirs.append("E")

    valid_dirs = [x for x in all_dirs if not any(y in invalid_dirs for y in x)]

    return [translate_step(x) for x in valid_dirs]


#Check the current provided location for the the target
def check_location(grid, loc, target="XMAS", first_char=True, set_dir=None):
    loc_value = grid["data"][loc[1]][loc[0]]

    #Check if potential target found
    if loc_value == target[0]:
        #If this was the end of the target, report the target was found
        if len(target) == 1:
            return 1
        #If the rest of the target needs to be found
        else:
            #If this is the first character then all directions need to be checked
            #There could be multiple valid targets from this starting point
            if first_char:
                #Get valid directions for the current location
                valid_dirs = get_valid_dirs(loc, grid)
                #Running total of matches found from this starting point
                matches_found = 0

                #Iterate through all valid directions
                for potential_dir in valid_dirs:
                    matches_found += check_location(
                        grid, 
                        apply_dir(loc, potential_dir), 
                        target[1:],
                        first_char=False, 
                        set_dir=potential_dir
                    )

                return matches_found
            #If not the first character than the current direction needs to be maintained
            else:
                return check_location(
                    grid,
                    apply_dir(loc, set_dir),
                    target[1:],
                    first_char=False, 
                    set_dir=set_dir
                )
    #Not a match
    else:
        return 0

#Function to traverse grid for words
def wordsearch(grid):

    #Running total
    words_found = 0
    
    #Iterate through the grid
    for idx_y in range(grid["height"]):
        for idx_x in range(grid["width"]):
            loc = (idx_x, idx_y)

            res = check_location(grid, loc)
            if res > 0:
                #print(f"{res} target(s) found at {loc}")
                pass

            words_found += res

    return words_found
```

```python
print(wordsearch(grid))
```
